```python
import pandas as pd
import matplotlib.pyplot as plt
# 노트북 안에 그래프를 그리기 위해
%matplotlib inline
import seaborn as sns
```

```python
train = pd.read_csv("data/train.csv", parse_dates=["datetime"])
train.shape
```

```python
train.info()
```

```python
train["year"] = train["datetime"].dt.year
train["month"] = train["datetime"].dt.month
train["day"] = train["datetime"].dt.day
train["hour"] = train["datetime"].dt.hour
train["minute"] = train["datetime"].dt.minute
train["second"] = train["datetime"].dt.second
train.shape
```

```python
train.head()
```

```python
train.dtypes
```

```python
train.isnull().sum()
```

```python
year=sns.barplot(data=train, x="year", y="count")
plt.savefig("./images/year.png")
```

```python
month=sns.barplot(data=train, x="month", y="count")
plt.savefig("./images/month.png")
```

```python
day=sns.barplot(data=train, x="day", y="count")
plt.savefig("./images/day.png")
```

```python
hour=sns.barplot(data=train, x="hour", y="count")
plt.savefig("./images/hour.png")
```

```python
sns.pointplot(data=train, x="hour", y="count")
plt.savefig("./images/pp-hour.png")
```

```python
sns.pointplot(data=train, x="hour", y="count", hue="workingday")
plt.savefig("./images/pp-hour-workingday.png")
```

```python
train["dayofweek"] = train["datetime"].dt.dayofweek
train.sample(5)
```

```python
sns.pointplot(data=train, x="hour", y="count", hue="weather")
plt.savefig("./images/pp-hour-weather.png")
```

```python
sns.pointplot(data=train, x="hour", y="count", hue="season")
plt.savefig("./images/pp-hour-season.png")
```
