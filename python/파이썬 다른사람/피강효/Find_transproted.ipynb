```python
data.isna().sum()
```

```python
data['Age_range']=pd.cut(data['Age'],bins=[-1,12,17,25,30,50,np.inf],labels=['until12','until17','until25','until30','until50','more then'])
paylist=['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
data['totalpay']=data[paylist].sum(axis=1)
data.loc[data['totalpay']>0,'spending']=0
data.loc[data['totalpay']==0,'spending']=1
data['Group']=data['PassengerId'].str.extract('(\d\d\d\d)\_',expand=False).astype(int)
data['Group_size']=data['Group'].map(dict(data['Group'].value_counts()))
data['Solo']=(data['Group_size']==1).astype(int)
data['Cabin1']=data.Cabin.str.extract('([A-Za-z]+)\/',expand=False)
data['Cabin2']=data.Cabin.str.extract('(\d\d\d\d|\d\d\d|\d\d|\d)')
data['Cabin3']=data.Cabin.str.extract('/([A-Za-z]+)',expand=False)
data['Cabin2']=pd.array(data['Cabin2'],dtype=pd.Int32Dtype())
data['Cabin2_cut']=pd.cut(data['Cabin2'],bins=[-1,300,600,900,1200,1500,1800,np.inf],labels=['300','600','900','1200','1500','1800','more then 1800'])
data['lastname']=data['Name'].str.extract('\ ([A-Za-z]+)',expand=False)
data['Family_size']=data['lastname'].map(dict(data['lastname'].value_counts()))
na_cols=data.columns[data.isna().any()].tolist()
na_cols.remove('Transported')
#data['na_count']=data.isna().sum(axis=1)
#data
```

- HomePlanet결측치 시작
- 결측치 288개

```python
#group에 때른 homePlanet 계산
before=data['HomePlanet'].isna().sum()#288
group_home=data.groupby(['Group','HomePlanet']).size()
groups_dict=dict(group_home.index)
data.loc[data['HomePlanet'].isna(),'HomePlanet']=data['Group'].loc[data['HomePlanet'].isna()].map(groups_dict)

print('before',before)
print('after',data['HomePlanet'].isna().sum())
```

```python

cabinhome=data.groupby(['Cabin1','HomePlanet'])['HomePlanet'].size()
#europa = a b c t
#earth = g
data.loc[(data['HomePlanet'].isna())&(data['Cabin1'].isin(['A','B','C','T'])),'HomePlanet']='Europa'
data.loc[(data['HomePlanet'].isna())&(data['Cabin1']=='G'),'HomePlanet']='Earth'
```

```python
#lastname에 따른 homeplanet 계산
before=data['HomePlanet'].isna().sum()#94
family=data.groupby(['lastname','HomePlanet']).size()
family_dict=dict(family.index)
data.loc[data['HomePlanet'].isna(),'HomePlanet']=data['lastname'].loc[data['HomePlanet'].isna()].map(family_dict)
print('before',before)
print('after',data['HomePlanet'].isna().sum())
```

```python
#destination과 관련지어 home추측
#cabin1==D 이면 mars  !=D이면 earth
#사이트 참조
data.loc[(data['HomePlanet'].isna())&(data['Cabin1']=='D'),'HomePlanet']='Mars'
data.loc[(data['HomePlanet'].isna())&(data['Cabin1']!='D'),'HomePlanet']='Earth'
print('after',data['HomePlanet'].isna().sum())
```

- 결측처리 중 이부분은 이해가 되지않아 그냥 사용함


```python
GCD_gb=data[data['Group_size']>1].groupby(['Group','Cabin1'])['Cabin1'].size().unstack().fillna(0)
GCN_gb=data[data['Group_size']>1].groupby(['Group','Cabin2'])['Cabin2'].size().unstack().fillna(0)
GCS_gb=data[data['Group_size']>1].groupby(['Group','Cabin3'])['Cabin3'].size().unstack().fillna(0)
```

```python
GCD_index=data[data['Cabin1'].isna()][(data[data['Cabin1'].isna()]['Group']).isin(GCD_gb.index)].index
data.loc[GCD_index,'Cabin1']=data.iloc[GCD_index,:]['Group'].map(lambda x: GCD_gb.idxmax(axis=1)[x])
```

```python
#각족 값의 join을 통해 cabin1에 대한 추적
data.groupby(['HomePlanet','Destination','Solo','Cabin1']).size().unstack().fillna(0)
index=data.loc[data['Cabin1'].isna(),'Cabin1'].index
data.loc[data['Cabin1'].isna(),'Cabin1']=data.groupby(['HomePlanet','Destination','Solo'])['Cabin1'].transform(lambda x:x.fillna(pd.Series.mode(x)[0]))[index]
data.loc[data['Cabin1'].isna(),'Cabin1']='F'
```
