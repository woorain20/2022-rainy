### 유니코드 에러 확인
 - UnicodeDecodeError 에러 발생
 - encoding="euc-kr" 추가로 해결

```python
df1=pd.read_csv("DataV2/Age.csv", encoding="euc-kr")
df1
```

```python
df2=pd.read_csv("DataV2/Gender.csv", encoding="euc-kr")
df2
```

```python
df3=pd.read_csv("DataV2/Size.csv", encoding="euc-kr")
df3
```

### concat을 통해 CSV 파일 하나로 합치기

```python
df4=pd.concat([df1,df2,df3], ignore_index=True)
df4
```

### 행과 열 바꾸기
- transpose() 사용

```python
df4=df4.transpose()
df4
```

```python
df4.info()
```
