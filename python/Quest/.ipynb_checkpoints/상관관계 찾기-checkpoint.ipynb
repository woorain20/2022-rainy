### 상관관계 확인

```python
corr=df9.corr()
corr
```

#### 결과 확인
 - 승차승객과 하차승객, 총 승객
 - 하차승객과 승차승객, 총 승객

 - 3개의 컬럼이 서로 상관관계를 갖고 있음

### 그래프 확인

```python
plt.figure(figsize=(14,8))
df9.corr()["총 승객"].sort_values()[:-1].plot(kind="bar")
plt.show()
```

```python
plt.figure(figsize=(14,8))
df9.corr()["하차승객"].sort_values()[:-1].plot(kind="bar")
plt.show()
```
