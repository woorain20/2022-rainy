```python
df.columns=['날짜', '노선', '역', '승차승객', '하차승객']
```

```python
df.insert(5, "총 승객", df["승차승객"]+df["하차승객"])
```

```python
df.insert(6, "승차-하차", df["승차승객"]-df["하차승객"])
```

```python
df
```

```python
df['날짜']= df['날짜'].astype('str')
```

```python
df['날짜'] = pd.to_datetime(df['날짜'])
```

```python
df['월']=df['날짜'].dt.month # 월 정보
df['일']=df['날짜'].dt.day # 일 정보
```

```python
df
```

```python
df['요일'] = df['날짜'].dt.weekday
```

```python
df
```

```python
df=df.drop("날짜", axis=1)
```

```python
df=df.drop("역", axis=1)
```

```python
line_mapping={"1호선":0, "2호선":1, "3호선":2, "4호선":3, "5호선":4, "6호선":5, "7호선":6, "8호선":7, "9호선":8, "9호선2~3단계":9, "경강선":10, "경부선":11, "경원선":12, 
              "경의선":13, "경인선":14, "경춘선":15, "공항철도 1호선":16, "과천선":17, "분당선":18, "수인선":19, "신림선":20, "안산선":21, "우이신설선":22,
              "일산선":23, "장항선":24, "중앙선":25}
```
