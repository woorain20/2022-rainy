# 수치형 변수

```python
import pandas as pd
import seaborn as sns
print(pd.__version__)
print(sns.__version__)
```

```python
# df -> pandas -> data frame
df=sns.load_dataset("mpg")
df
```

```python
df.info()
```

```python
# 히스토그램 - 표로 되어 있는 도수 분포
df.hist()
```

```python
# 글자가 잘 안 보여서 크기 조정
df.hist(figsize=(10,8))
```

```python
# bins 조정하면 데이터 범위가 달라져서 세부범위 확인 가능(bins 숫자 클수록 세세한 범위 확인 가능)
df.hist(figsize=(10,8), bins=60)
```

```python
# 비대칭도(왜도)
# 왜도가 음수일 경우에는 오른쪽으로 기운다
# 왜도가 양수일 경우에는 왼쪽으로 기운다
# 평균과 중앙값이 같으면 0
# skew - 왜도 구하기
df.skew()
```

```python
# 첨도
# 확률 분포의 뾰족한 정도를 나타내는 척도
# 데이터가 어느 곳에 집중적으로 몰려 있는가를 측정할 때 사용
# 첨도값이 3에 가까울수록 정규분포에 가깝다
# 첨도 < 3 더 뾰족한 분포(꼬리가 얇다) 
# 첨도 > 3 더 완만하고 납작한 분포(꼬리가 두꺼운 분포)
# kurt() - 첨도 구하기
df.kurt()
```

```python
# seaborn -> 히스토그램 -> displot()
sns.displot(data=df, x="mpg")
```

```python
sns.displot(data=df, x="mpg", kde=True)
# kde - 선을 그려줌
```

```python
# kdeplot() - 선으로 밀도 함수 표현
sns.kdeplot(data=df, x="mpg")
```

```python
sns.kdeplot(data=df, x="mpg", shade=True)
# shade - 색 표시
```

```python
# mpg 값에 대해 skew, kurt 값 구하기
# 하나의 칼럼 객체 -> 시리즈(Serise)
df["mpg"]
```

```python
df["mpg"].agg(["skew", "kurt"])
```

```python
# boxplot으로 mpg의 사분위 수 시각화
sns.boxplot(data=df, x="mpg")
```

```python
sns.boxplot(data=df, x="weight")
```
