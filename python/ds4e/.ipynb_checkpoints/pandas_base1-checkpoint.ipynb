```python
df1=sns.load_dataset("mpg")
df1.info()
```

```python
# 데이터가 빠졌다 - 결측치
```

```python
df1.isnull()
# data가 있으면 false 없으면 true
```

```python
print(int(True), int(False))
print(True+False+True)
```

```python
df1.isnull().sum()
# 결측치가 몇 개 있는지 파악 가능
```

```python
# 결측치 비율 확인
# mean() - 평균값 구하는 함수
df1.isnull().mean()
```

```python
# describe() - 기술 통계
df1.describe()
```

```python
# std - 표준편차
```

```python
# object의 기술 통계 보기 위해서는
df1.describe(include="object")
```
